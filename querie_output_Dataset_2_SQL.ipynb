```
%sql
-- 01) Liste a quantidade de pagamentos por professional_status (tabelas: funnel e customers)

SELECT COUNT(sales.funnel.paid_date) AS total_pagt , sales.customers.professional_status AS status_profissional
FROM sales.funnel
INNER JOIN sales.customers ON sales.customers.customer_id = sales.funnel.customer_id
GROUP BY sales.customers.professional_status;

```

```
%sql
-- 02) Liste a quantidade de visitas por marca de veiculo (tabelas: funnel e products )

SELECT COUNT(sales.funnel.visit_id) AS QuantidadeVisitas, sales.products.brand AS Fabricante
FROM sales.funnel
INNER JOIN sales.products ON sales.funnel.product_id = sales.funnel.product_id
GROUP BY brand
ORDER BY brand DESC;
```

```
%sql
-- 03) Liste quantos clientes moram em cada tamanho de cidade. O porte da cidade consta na coluna "size" da tabela temp_tables.regions (tabelas: customers e regions)

SELECT customers.city AS Porte_Cidade, customers.customer_id Total_Clientes
FROM sales.customers
INNER JOIN temp_tables.regions ON sales.customers.city = temp_tables.regions.city
UNION
SELECT DISTINCT size, COUNT(temp_tables.regions.city)
FROM temp_tables.regions
GROUP BY regions.size
```

```
%sql
-- 04) Liste a quantidade de visitas por lojas de ve√≠culo (tabelas: funnel e stores)

SELECT COUNT(funnel.visit_id) AS Visitantes ,stores.store_name AS Loja
FROM sales.funnel
LEFT JOIN sales.stores ON sales.funnel.store_id = sales.stores.store_id
GROUP BY stores.store_name
ORDER BY Visitantes DESC
```

```
%sql
-- 05) Liste a quantidade de pagamentos porgenero (tabelas: funnel, customers, ibge_genders)

SELECT COUNT(DISTINCT sales.funnel.paid_date) AS Qtd_pgtos, temp_tables.ibge_genders.gender, sales.customers.customer_id
FROM sales.funnel
JOIN sales.customers ON sales.customers.customer_id = sales.funnel.customer_id 
JOIN temp_tables.ibge_genders ON UPPER(temp_tables.ibge_genders.first_name) = sales.customers.first_name
WHERE paid_date IS NOT NULL
AND temp_tables.ibge_genders.gender = 'male'
GROUP BY temp_tables.ibge_genders.gender, sales.funnel.paid_date, sales.customers.customer_id
```
